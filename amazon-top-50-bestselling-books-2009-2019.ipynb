```python
#getting the data set
df = pd.read_csv("../input/amazon-top-50-bestselling-books-2009-2019/bestsellers with categories.csv")

#getting the column names
print('Columns')
print(df.columns)
print('\n')
print('----------------')


#rename the column name for the title from 'Name' to 'Title'
df.rename(columns={'Name': 'BookTitle'}, inplace=True)

#getting the top rows
print('Top 5 rows')
print(df.head())
print('\n')
print('----------------')

#getting the total number of books
booksum = df['BookTitle'].count()
print("Total number of Books:", booksum)
print('\n')
print('----------------')

#getting the unique number of books
booksum_unique = df['BookTitle'].nunique()
print("Unique number of Books:", booksum_unique)
print('\n')
print('----------------')

#getting the total number of authors
authorsum = df['Author'].count()
print("Total number of Authors:", authorsum)
print('\n')
print('----------------')

#getting the unique number of authors
authorsum_unique = df['Author'].nunique()
print("Unique number of Authors:", authorsum_unique)
print('\n')
print('----------------')

#getting the total count of books by genre
booksbygenre = df['Genre'].value_counts()
print(booksbygenre)
print('\n')
print('----------------')

#getting the total count of authors by genre
genregroup = df.groupby('Genre')['Author'].nunique()
print(genregroup)
print('\n')
print('----------------')

#top 10 authors with the highest reviews
sortauthors = df.sort_values(['Reviews'])
groupauthors = sortauthors.groupby('Author')['Reviews']
sumreviews_a= groupauthors.sum().sort_values(ascending=False)
topnauthor_reviews = sumreviews_a.head(10)
print(topnauthor_reviews)
print('\n')
print('----------------')



#top 10 books with the highest reviews
sortreviews = df.sort_values(['Reviews'])
groupreviews = sortreviews.groupby('BookTitle')['Reviews']
sumreviews= groupreviews.sum().sort_values(ascending=False)
topnbook_reviews = sumreviews.head(10)
print(topnbook_reviews)
```
